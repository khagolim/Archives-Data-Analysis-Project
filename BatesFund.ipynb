```r
#Graphs the overall donations to Bates from 1940 to 2020. The line of best fit shows that there is a strong negative correlation between overall giving & class year
bf1 <- ggplot(bf_data,aes(Class_Year, Overall_Giving)) + geom_point() + geom_smooth(method='lm')
bf1 + scale_y_continuous(trans='log10')
```

```r
# Open a pdf file
jpeg("Total_Donations_vs_Class_Year.jpeg", units="in", width=5, height=5, res=300) 
# 2. Create the plot
bf1 <- ggplot(bf_data,aes(Class_Year, Overall_Giving)) + geom_point() + geom_smooth(method='lm')
bf1 + scale_y_continuous(trans='log10')
# 3. Close the file
dev.off()
```

```r
#Graphs the percentage of female donors from the 1800s who donated to Bates, and the overall amount they gave to Bates
fm_donors_1800 <- ggplot(data = data_1) + geom_point(mapping = aes(x = Female.Donors., y = Overall.Giving))
fm_donors_1800 + scale_y_continuous(trans='log10')
```

```r
#Graphs the percentage of female donors from the 1900s who donated to Bates, and the overall amount they donated to Bates
fm_donors_1900 <- ggplot(data = data_2) + geom_point(mapping = aes(x = Female.Donors., y = Inflation.Adjustment.Amount))
fm_donors_1900 
```

```r
d1 <- data_1[c("Female.Donors.", "Overall.Giving", "Year.Donated", "year_num")]
#creates a data set with selected columns
d2 <- data_2[c("Female.Donors.", "Inflation.Adjustment.Amount", "Class_Year", "year_num")]
#creates a data set with selected columns

d1["century"] <- "19th"
#adds a new column to the newly created dataset
d2["century"] <- "20th"
#adds a new column to the newly created dataset

names(d1)[names(d1) == "Overall.Giving"] <- "Inflation.Adjustment.Amount"
#creates a matched column to help in the binding process
names(d1)[names(d1) == "Year.Donated"] <- "Class_Year"
#creates a matched column to help in the binding process
d3 <- rbind(d1,d2)
d3
```

```r
ggplot(data=d3, aes(x=year_num, y=Female.Donors., fill=century)) + geom_bar(stat="identity", position=position_dodge())
```

```r
# Open a pdf file
jpeg("Female_Donors_19th_vs_20th_Century.jpeg", units="in", width=5, height=5, res=300) 
# 2. Create the plot
ggplot(data=d3, aes(x=year_num, y=Female.Donors., fill=century)) + geom_bar(stat="identity", position=position_dodge())
# 3. Close the file
dev.off()
```

```r
#Bar graph showing the inflation adjusted amount donated to Bates over the 5 year period from 1854-1858 and 1954-1958 
ggplot(data=d3, aes(x=year_num, y=Inflation.Adjustment.Amount, fill=century)) + geom_bar(stat="identity", position=position_dodge())
```

```r
# Open a pdf file
jpeg("Total_Donations_19th_vs_20th_Century.jpeg", units="in", width=5, height=5, res=300) 
# 2. Create the plot
ggplot(data=d3, aes(x=year_num, y=Inflation.Adjustment.Amount, fill=century)) + geom_bar(stat="identity", position=position_dodge())
# 3. Close the file
dev.off()
```
